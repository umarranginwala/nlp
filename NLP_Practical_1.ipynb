###Removing Punctuations from the text
One of the common techniques in text preprocessing is the removal of punctuation from textual data. This process standardizes the text and treats words like 'hurray' and 'hurray!' equally.

It's important to thoughtfully select which punctuation marks to exclude, depending on the specific use case. For example, the string.punctuation module in Python encompasses various punctuation symbols:

!"#$%&'()*+,-./:;<=>?@[\]^_{|}~

You have the flexibility to modify this list of punctuation marks according to your requirements. Let's illustrate this technique with an example from a product review tweet dataset:

Example
Happiness in a box!!! Yey!!! #me #phone #sony #xperia #sonyxperiaZ #new #blessed… http://instagram.com/p/exAJ7tloSM/


By removing punctuation, the tweet becomes:

Happiness in a box Yey me phone sony xperia sonyxperiaZ new blessed httpinstagramcompexAJ7tloSM

```python
print(string.punctuation)
```

```python
data.head(5)
```

```python
punctuation_to_remove = string.punctuation
```

```python
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', punctuation_to_remove))

data["tweet_without_punctuation"] = data["tweet_lower"].apply(lambda text: remove_punctuation(text))
data.head(5)
```

###Removal of stopwords

Stopwords are those common words we use all the time, like 'the' or 'a'. Usually, we skip them when working with text because they don't add much useful information for later analysis. Stopwords are words that don't contribute much to the overall meaning of a text. In English, these could include words like "this," "is," "the," "it," and "so." These words often appear frequently across texts, and when analyzing text, they can sometimes overshadow more important words that carry significant meaning.

Here we are deadling with a dataset of product review tweets, if we have a tweet: "This phone is the best, I love it so much!"



In the example tweet, if we remove stopwords, we might end up with: "phone best, love much!" The words "this," "is," "the," and "it" have been removed because they are stopwords. However, the remaining words like "phone," "best," "love," and "much" convey the key sentiment and information of the tweet.

```python
nltk.download('stopwords')

from nltk.corpus import stopwords
print(stopwords.words('english'))
```

```python
# Define a function to remove stopwords
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """Custom function to remove stopwords"""
    tokens = text.split()
    filtered_tokens = [word for word in tokens if word.lower() not in STOPWORDS]
    return ' '.join(filtered_tokens)
```

```python
data["tweet_without_stop_words"] = data["tweet_without_punctuation"].apply(remove_stopwords)
```

```python
data.head()
```

### Removal of Frequent words

In the preceding preprocessing phase, we eliminated stopwords by considering linguistic characteristics. However, in scenarios where we possess a corpus specific to a particular domain, there could exist frequently occurring words that hold lesser significance for our analysis.

```python
#Let's identify the most prevalent words and proceed to eliminate them in the subsequent phase.
```

```python
from collections import Counter
cnt = Counter()
for text in data["tweet_without_stop_words"].values:
    for word in text.split():
        cnt[word] += 1

cnt.most_common(10)
```

```python
all_words = ' '.join(data["tweet_without_stop_words"]).split()
cnt = Counter(all_words)
FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
FREQWORDS
```

```python
def remove_freqwords(text):
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])
```

```python
data["tweet_without_frequent_words"] = data["tweet_without_stop_words"].apply(remove_freqwords)
```

```python
data.head()
```

## Removal of Rare words

```python
# Define the number of rare words to keep
num_rare_words = 10

# Create a set of rare words based on frequency
all_words = ' '.join(data["tweet_without_stop_words"]).split()
word_count = Counter(all_words)
rare_words_set = set([word for (word, word_count) in word_count.most_common()[:-num_rare_words-1:-1]])

# Define a function to remove rare words
def remove_rare_words(text):
    """Custom function to remove rare words"""
    words = text.split()
    filtered_words = [word for word in words if word not in rare_words_set]
    return ' '.join(filtered_words)

# Apply the function to the 'tweet_without_stop_words' column and create a new column 'tweet_without_frequent_rare_words'
data["tweet_without_frequent_rare_words"] = data["tweet_without_stop_words"].apply(remove_rare_words)

# Display the updated DataFrame
data.head()
```

##URL Removal

The next step in preprocessing entails the removal of any URLs present within the product review dataset.

If we're analyzing product reviews, it's possible that some reviews may include URLs. For the sake of our comprehensive analysis, it might be prudent to exclude these URLs.

```python
# Define a function to remove URLs using str.replace()
def remove_urls(text):
    return re.sub(r'http\S+|www\S+', '', text)

# Apply the function to the 'tweet' column and create a new column 'tweet_without_urls_alt'
data["tweet_without_urls"] = data["tweet_without_frequent_rare_words"].apply(remove_urls)

# Display the updated DataFrame
data.head()
```

# Stemming


Stemming involves the process of simplifying inflected or derived words to their fundamental word stem, base, or root form. This procedure aims to achieve word reduction while maintaining the core meaning of words within a given context.

Various stemming algorithms are available, each offering distinct methodologies. One notable approach is the Porter stemmer, widely recognized and embraced within linguistic applications. Leveraging tools like the NLTK package enables the seamless integration of stemming processes for enhanced text analysis.

```python
from nltk.stem.porter import PorterStemmer
# Initialize the PorterStemmer
stemmer = PorterStemmer()

# Define a function to stem words
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

# Apply the function to the 'text' column and create a new column 'text_stemmed'
data["tweet_stemmed"] = data["tweet_without_urls"].apply(stem_words)

# Display the updated DataFrame
data.tail()
```

## Lemmatization
Lemmatization is a text preprocessing technique in natural language processing (NLP) that involves reducing words to their base or root form, known as a "lemma." The goal of lemmatization is to transform different inflected forms of a word into a common base form so that they can be treated as the same word during analysis. This process helps in standardizing and simplifying words while retaining their essential meaning.

Unlike stemming, which involves removing prefixes and suffixes from words to obtain their root form, lemmatization ensures that the resulting lemma is a valid word in the language. This linguistic validation is achieved by using a dictionary or vocabulary (such as WordNet in NLTK) to map words to their lemmatized forms.

For example:

Lemmatization of the word "running" might result in the lemma "run."
Lemmatization of the word "better" might result in the lemma "good."

```python
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

# Download NLTK WordNet and POS tagger data
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')


# Initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Map POS tags to WordNet POS tags
wordnet_map = {"N": wordnet.NOUN, "V": wordnet.VERB, "J": wordnet.ADJ, "R": wordnet.ADV}

# Define a function to lemmatize words with POS tagging
def lemmatize_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])

# Apply the function to the 'tweet' column and create a new column 'text_lemmatized'
data["text_lemmatized"] = data["tweet_without_urls"].apply(lemmatize_words)

# Display the updated DataFrame
data.tail()
```
